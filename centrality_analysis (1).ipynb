```python
import networkx as nx
import matplotlib.pyplot as plt

```

```python
G_florentine = nx.florentine_families_graph()
betweenness = nx.betweenness_centrality(G_florentine)
```

```python
plt.figure(figsize=(12, 8))
node_size = [b * 5000 for b in betweenness.values()]
node_color = ['red' if node in ['Medici', 'Pazzi'] else 'blue' for node in G_florentine.nodes()]

nx.draw(
    G_florentine,
    with_labels=True,
    node_size=node_size,
    node_color=node_color,
    edge_color='gray'
)

plt.title("Floransa Family")
plt.show()
```
