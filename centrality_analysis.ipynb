```python
from google.colab import files
import networkx as nx
import matplotlib.pyplot as plt



```

```python

uploaded = files.upload()

G_football = nx.read_pajek(list(uploaded.keys())[0])
G_football = G_football.to_undirected()

```

```python

pagerank = nx.pagerank(G_football)
betweenness = nx.betweenness_centrality(G_football)

```

```python

plt.figure(figsize=(15, 10))
ax = plt.gca()

nx.draw(
    G_football,
    ax=ax,
    with_labels=True,
    node_size=[p * 5000 for p in pagerank.values()],
    node_color=list(betweenness.values()),
    cmap=plt.cm.plasma,
    edge_color='gray'
)

sm = plt.cm.ScalarMappable(
    cmap=plt.cm.plasma,
    norm=plt.Normalize(vmin=min(betweenness.values()), vmax=max(betweenness.values()))
)
plt.colorbar(sm, ax=ax, label="Betweenness")
plt.show()
```
